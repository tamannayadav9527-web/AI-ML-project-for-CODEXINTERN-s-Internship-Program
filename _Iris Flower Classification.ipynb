```python
df=df[df['species']!='Iris-setosa']
df['species'].value_counts()
```

```python
df['species']=df['species'].map({"Iris-versicolor":1,"Iris-virginica":0})
```

```python
X = df.drop("species", axis=1)
```

```python
Y = df["species"]  
```

```python
# Split the data into training/test sets:

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.6,random_state=55)
print(x_train.shape, x_test.shape)
```

```python
#Application of the model into the dataset:

from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
print (classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_pred,y_test)
```
